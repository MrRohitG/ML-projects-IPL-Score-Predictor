```python
import pandas as pd
import numpy as np
from scipy.stats import pearsonr, spearmanr, kendalltau
from scipy.stats import ttest_1samp, ttest_ind, ttest_rel, chi2_contingency, f_oneway, friedmanchisquare

# Load the dataset
# Assuming `data.csv` is your IPL dataset
data = pd.read_csv("ipl_data.csv")

# 1. Descriptive Statistics
print("Descriptive Statistics:")
print(data.describe())  # Summary statistics for numerical columns

# 2. Data Preprocessing
# Fill missing values for numeric columns with their mean
numeric_data = data.select_dtypes(include=['float64', 'int64'])
data[numeric_data.columns] = data[numeric_data.columns].fillna(numeric_data.mean())

# Fill missing values for categorical columns with the mode
categorical_data = data.select_dtypes(include=['object'])
data[categorical_data.columns] = data[categorical_data.columns].fillna(categorical_data.mode().iloc[0])

# Convert categorical columns to numerical using one-hot encoding
data = pd.get_dummies(data, drop_first=True)

# 3. Pearson's Correlation Coefficient
print("\nPearson's Correlation Coefficient:")
pearson_corr, _ = pearsonr(data['runs'], data['wickets'])  # example correlation between runs and wickets
print("Pearson correlation (runs vs wickets):", pearson_corr)

# 4. Spearman's Correlation Coefficient
print("\nSpearman's Correlation Coefficient:")
spearman_corr, _ = spearmanr(data['runs'], data['wickets'])
print("Spearman correlation (runs vs wickets):", spearman_corr)

# 5. Kendall Tau's Correlation Coefficient
print("\nKendall Tau's Correlation Coefficient:")
kendall_corr, _ = kendalltau(data['runs'], data['wickets'])
print("Kendall Tau correlation (runs vs wickets):", kendall_corr)

# 6. T-tests
# (i) One-sample t-test: Test if mean of runs is different from a value (e.g., 10)
print("\nOne-sample T-test:")
t_stat, p_value = ttest_1samp(data['runs'], 10)
print("One-sample t-test (mean runs == 10): t-statistic =", t_stat, ", p-value =", p_value)

# (ii) Independent-sample t-test: Test if two independent groups (e.g., runs in 2 different matches) differ in mean
print("\nIndependent-sample T-test:")
t_stat, p_value = ttest_ind(data[data['mid'] == 1]['runs'], data[data['mid'] == 2]['runs'])
print("Independent t-test (match 1 vs match 2): t-statistic =", t_stat, ", p-value =", p_value)

# (iii) Paired T-test: Test if the difference in two related columns (e.g., runs and runs_last_5) is significant
print("\nPaired T-test:")
t_stat, p_value = ttest_rel(data['runs'], data['runs_last_5'])
print("Paired t-test (runs vs runs_last_5): t-statistic =", t_stat, ", p-value =", p_value)

# 7. Chi-Squared Test
# Test independence between two categorical variables (e.g., team name and win/loss outcome)
print("\nChi-Squared Test:")
# Example with dummy variables 'bowler_WPUJC Vaas' and 'bowler_A Sharma'
contingency_table = pd.crosstab(data['bowler_WPUJC Vaas'], data['bowler_A Sharma'])
chi2, p, dof, expected = chi2_contingency(contingency_table)
print("Chi-squared test: chi2 =", chi2, ", p-value =", p)

# 8. F-Test (One-way ANOVA)
# Compare means of runs across multiple matches
print("\nF-Test (One-way ANOVA):")
anova_result = f_oneway(data[data['mid'] == 1]['runs'], data[data['mid'] == 2]['runs'], data[data['mid'] == 3]['runs'])
print("One-way ANOVA (runs across 3 matches): F-statistic =", anova_result.statistic, ", p-value =", anova_result.pvalue)

# 9. Friedman Test
# Compare rankings of a variable (e.g., runs) across different conditions (e.g., last few overs in different matches)
print("\nFriedman Test:")
# Sample data from 3 overs for 5 matches
friedman_result = friedmanchisquare(data[data['mid'] == 1]['runs'][:5], data[data['mid'] == 2]['runs'][:5], data[data['mid'] == 3]['runs'][:5])
print("Friedman test (runs over 3 matches): chi2 =", friedman_result.statistic, ", p-value =", friedman_result.pvalue)
```
